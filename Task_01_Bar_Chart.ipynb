```python
import zipfile
import pandas as pd
import matplotlib.pyplot as plt
zip_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.zip'
csv_filename = 'API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv'


with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    with zip_ref.open(csv_filename) as csvfile:
        df = pd.read_csv(csvfile, skiprows=4)
india_data = df[df['Country Name'] == 'India']
years = india_data.columns[4:-1]
population = india_data[years].values.flatten()
pop_df = pd.DataFrame({
    'Year': years.astype(str),
    'Population': population.astype(float)
})
plt.figure(figsize=(14, 6))
plt.bar(pop_df['Year'], pop_df['Population'], color='brown')
plt.xticks(rotation=90)
print("                        ------->INTERNSHIP Task-01 Completed Successfully.<---------")
print("\n")
plt.title('India Population Distribution Over Years 1960-2024')
plt.xlabel('Year')
plt.ylabel('Population')
plt.tight_layout()
plt.savefig('india_population_bar_chart.png')
plt.show()

print(" This bar chart visualizes the change in India’s population from 1960 to 2024.")
```
